```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
x_data = pd.read_excel("./result/x_move_0.1.xlsx", header=0)

plt.figure()
plt.title("Box-Whisker Plot")
plt.rcParams["figure.figsize"] = (25,8)
plt.rcParams["axes.titlesize"] = 25
plt.rcParams["axes.labelsize"] = 15
plt.ylabel("Relative Position(mm)")
plt.xlabel("Relative Position(mm)")
plt.legend()
plt.boxplot([x_data['ID30-X0'], x_data['ID30-X1']])
plt.xticks([1, 2], ['X0', 'X1'])
plt.show()

print("X0 mean : {}".format(x_data['ID30-X0'].mean()))
print("X1 mean : {}".format(x_data['ID30-X1'].mean()))

```

```python
x_data = pd.read_excel("./result/x_move_0.1.xlsx", header=0)

plt.figure()
plt.title("Reference Wafer Transfer Precision")
plt.rcParams["figure.figsize"] = (8,8)
plt.rcParams["axes.titlesize"] = 20
plt.rcParams["axes.labelsize"] = 15
plt.ylabel("Relative Position(mm)")
plt.xlabel("Wafer Transfer Point (x0=start, x1=x0+0.1mm)")
plt.legend()
plt.boxplot([x_data['x0'], x_data['x1']], showmeans=True)
plt.xticks([1, 2], ['X0', 'X1'])
plt.show()

print("X0 mean : {}, std : {}".format(x_data['x0'].mean(), x_data['x0'].std()))
print("X1 mean : {}, std : {}".format(x_data['x1'].mean(), x_data['x1'].std()))
```
